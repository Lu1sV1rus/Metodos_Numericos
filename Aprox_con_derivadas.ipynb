```python
def fun(x):
  g1=x-(x**3)-(4*(x**2))+10
  return g1

#Tolerancia
ep=1e-5
#Rango
Pmin=1
Pmax=2
#Contador y funciones
P0=Pmin
fx=fun(P0)
N=1

print("N  | P0 | g(P0) | P0-g(P0)")           #Formato de Tabla
print(N," | ",P0," | ",fx," | ",abs(P0-fx))   #Primer valor
while abs(P0-fx)>ep:                          #Mientras sea mayor al error
  if P0<Pmax:                                 #Revisamos que se mantenga en el rango
    N=N+1                                     #Actualizamos el contador
    P0=fx                                     #Actualizamos P0
    fx=fun(P0)                                #Evaluamos en el punto
  else:                                       #En caso de salir de rango, no va a converger
    print("El valor P0 =",P0," se salio del rango [",Pmin,",",Pmax,"]")
    break
  print(N," | ",P0," | ",fx," | ",abs(P0-fx))
print("Punto fijo = ",P0)
print("Punto valuado = ",fun(P0))
print("Valor de e = ",abs(P0-fx))
print("Numero de iteraciones",N)
```
