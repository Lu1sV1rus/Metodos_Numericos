```python
#Epsilon
ep=1e-3           #El valor de ep minimo, para saber que se acerca al Punto Fijo
epmax=1e10        #El valor de ep maximo, para saber que se va al infinito

Pmin=1            #Donde inicia P0

Nit=10            #Numero de iteraciones
```

```python
def fun(x):
  g1=x-(x**3)-(4*(x**2))+10
  return g1

#Funciones
P0=Pmin
fx=fun(P0)

print("N | P0 | g(P0) | |P0-g(P0)|")
print("1 | {} | {} | {}".format(P0,fx,abs(P0-fx)))
for i in range(1,Nit):                      #Realizara iteraciones hasta Nit
  if abs(P0-fx)<epmax:                      #Mientras este en el rango de e
    if abs(P0-fx)>ep:
      P0=fx                                 #Realizara la sustitucion de P0
      fx=fun(P0)                            #Evaluara P0
    else:
      print("-----------------------------")      #En caso de llegar antes a e
      print("La ecuacion a llegado a e, antes")   # termina en programa
      break
  else:
    print("-----------------------------")        #En caso de salirse del rango max
    print("La ecuacion parece no converger")      # termina en program
    break
  print("{} | {} | {} | {}".format(i+1,P0,fx,abs(P0-fx)))
print("-----------------------------")            #Imprime los resultados
print("Punto fijo {}".format(P0))
print("Punto valuado = {}".format(fun(P0)))
print("Diferncia p = {}".format(abs(P0-fx)))
print("Numero de iteraciones",i)
```
