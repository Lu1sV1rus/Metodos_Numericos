# Algoritmo para el metodo de Biseccion

Entrada: Limite inferior $a$, Limite superior $b$, $\epsilon$ tolerancia, $N_{i}$ numero de iteraciones.

Paso 1.- Selecionemos $a$ y $b$ tales que $f(a)f(b) < 0$.\

Paso 2.- Tomar $N_{i} = 0$, $FA = f(a)$.\

Paso 3.- definamos $p = (a+b)/2$ y evaluemos $FP = f(p)$

Paso 4.- Mientras $\left|\frac{b-a}{2}\right|\geq\epsilon$ repetir los pasos 5 - 7 \

Paso 5.- Si $FA*FP > 0$ reemplazar $a = p$.
  En caso contrario $b = p$. \

Paso 6.- calcular p = (a+b)/2, $FP = f(p)$.

Paso 7 .- Incrementar el contador $N_{i} = N_{i} + 1$

Salida $p$ es la aproximacion a la raiz, $N_{i}$ numero total de iteraciones.






```python
# Programa del metodo de biseccion

import math
import matplotlib.pyplot as plt

def f(x):
    fac1 = pow(x,2)
    fac2 = -4.0
    fac3 = fac1 + fac2
    return fac3

# dominio de la funcion
xmin = -5.0
xmax = 5.0
Nx = 2**8
dx = (xmax-xmin)/(Nx-1)
lx = []
lf = []
for ix in range(Nx):
    xi = xmin + ix*dx
    fi = f(xi)
    lx.append(xi)
    lf.append(fi)
```

```python
plt.figure(figsize=(3,3))
plt.plot(lx,lf)
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

```python
# Metodo de biseccion
epsilon = 1.0e-10
Ni = 0.0
a = float(input("Intruzca el valor de a"))
b = float(input("Intruzca el valor de b"))
FA = f(a)
p = 0.5*(a+b)
FP = f(p)

#while abs(0.5*(b-a)) >= epsilon:
# La condicion anterior no del todo es
# la adecuada.

Ni = 0.0 # Contador inicial
while abs(FP) >= epsilon:
    if FA*FP > 0.0:
        a = p
    else :
        b = p
    p = 0.5*(a+b)
    FP = f(p)
    Ni += 1.0
print("El numero de iteraciones es N = {}".format(Ni))
print("El valor aproximado de la raiz es p = {}".format(p))
```
