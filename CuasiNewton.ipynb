```python
#2x2
import sympy as sp
import math
from sympy import symbols, Eq, solve
x, y = symbols("x y")
#Funciones
f1=-1+3*x**2-2*y**2
f2=3*x*y**2-y**3-1
M=[[(sp.diff(f1,x)),(sp.diff(f1,y))],[(sp.diff(f2,x)),(sp.diff(f2,y))]]
print("")
print(M)
print("")
```

```python
def eva(i,x,y):
  g1=-1+3*x**2-2*y**2
  g7=3*x*y**2-y**3-1
  g3=0
  if i==1:
    return float(g1)
  if i==2:
    return float(g7)

def eva2(M,i,j):
  v1=M[i][j]
  a=lambda x,y:v1
  v=a(x,y)
  return float(v)
```

```python
#2x2
import numpy as np
#Intervalo
x=0.5
y=0.75
#Numero de iteraciones
iii=4
#No olvidar Matriz
for i in range(iii):
  #Matriz
  M=[[6*x, -4*y], [3*y**2, 6*x*y - 3*y**2]]
  M3=np.array([eva(1,x,y),eva(2,x,y)])
  M2=np.array([[eva2(M,0,0), eva2(M,0,1)], [eva2(M,1,0), eva2(M,1,1)]])
  B1=np.linalg.inv(M2)
  X=B1.dot(M3)
  x=x-X[0]
  y=y-X[1]
  print("J",i+1)
  print(M2)
  print("J-1=F(x)")
  print(B1,"=",M3)
  print("h=",X[0]," k=",X[1])
  print("x=",x," y=",y)
  print("error= ",((X[0])**2+(X[1])**2)**(1/2))
  print("------------------------------------")
  M=[]
  M3=[]
  M2=[]
  B1=[]
  X=[]
```

```python
#3x3
import sympy as sp
import math
from sympy import symbols, Eq, solve
x, y, z = symbols("x y z")
f1=2*x-3*y+z-4
f2=2*x+y-z+4
f3=x**2+y**2+z**2-4
M=[[(sp.diff(f1,x)),(sp.diff(f1,y)),(sp.diff(f1,z))],[(sp.diff(f2,x)),(sp.diff(f2,y)),(sp.diff(f2,z))],[(sp.diff(f3,x)),(sp.diff(f3,y)),(sp.diff(f3,z))]]
print("")
print(M)
print("")
```

```python
def eva3x3(i,x,y,z):
  g1=2*x-3*y+z-4
  g7=2*x+y-z+4
  g3=x**2+y**2+z**2-4
  if i==1:
    return float(g1)
  if i==2:
    return float(g7)
  if i==3:
    return float(g3)
def eva3(M,i,j):
  v1=M[i][j]
  a=lambda x,y,z:v1
  v=a(x,y,z)
  return float(v)
```

```python
#3x3
import numpy as np
#Intervalo
x=-0.5
y=-1.5
z=1.5
#Numero de iteraciones
iii=4
#No olvidar la Matriz
for i in range(iii):
  #Matriz
  M=[[2, -3, 1], [2, 1, -1], [2*x, 2*y, 2*z]]
  M3=np.array([eva3x3(1,x,y,z),eva3x3(2,x,y,z),eva3x3(3,x,y,z)])
  M2=np.array([[eva3(M,0,0), eva3(M,0,1), eva3(M,0,2)], [eva3(M,1,0), eva3(M,1,1), eva3(M,1,2)],[eva3(M,2,0), eva3(M,2,1), eva3(M,2,2)]])
  B1=np.linalg.inv(M2)
  X=B1.dot(M3)
  x=x-X[0]
  y=y-X[1]
  z=z-X[2]
  print("J",i+1)
  print(M2)
  print("J-1=F(x)")
  print(B1,"=",M3)
  print("h=",X[0]," j=",X[1]," k=",X[2])
  print("x=",x," y=",y," z=",z)
  print("error= ",((X[0])**2+(X[1])**2+(X[2])**2)**(1/2))
  print("------------------------------------")
  M=[]
  M3=[]
  M2=[]
  B1=[]
  X=[]
```
