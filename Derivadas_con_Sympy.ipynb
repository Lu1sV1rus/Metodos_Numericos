# Ejemplos de calculo con python.

Determine. $max _{a\leq x \leq b}|f(x)|$ donde $f(x) = 5\cos \left(2x\right) - 2x\sin\left(2x\right)$ en los intervalos $[1,2]$ y $[0.5,1]$

```python
# Esto es un comentario
# importaremos la libreria sympy
import sympy as sp

# declararemo las siguientes variables como
# variables simbolicas
x , y ,z = sp.symbols('x y z')

# Definicion de una funcion simbolica
f = 5*sp.cos(2*x) -2*x*sp.sin(2*x)
f1 = sp.diff(f,x)
f2 = sp.diff(f1,x)

# Definamos una funcion de punto flotante
fx = sp.lambdify(x,f)
f1x = sp.lambdify(x,f1)
```

```python
# Obtendremos la raiz de f'(x)
# o bien el punto critico de f(x)
xc = float(sp.nsolve(f1,x,[1.0,2.0]))
display(xc)
```

```python
# Grficar la funcion fx
import matplotlib.pyplot as plt

x_min = 0.5
x_max = 1.0
N_x = pow(2,8)
d_x = (x_max - x_min)/(N_x - 1)
l_x = [] # esta es un alista vacia
l_fx = []
l_f1x = []
for ix in range(N_x):
    xi = x_min + ix*d_x
    fxi = fx(xi)
    f1xi = f1x(xi)
    l_x.append(xi)
    l_fx.append(fxi)
    l_f1x.append(f1xi)

plt.figure(figsize = (3,3))
plt.plot(l_x,l_fx)
plt.plot(l_x,l_f1x)
plt.grid(True)
plt.title('Grafica de la funcion')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
