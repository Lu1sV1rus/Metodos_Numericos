```python
#2x2
import sympy as sp
import math
from sympy import symbols, Eq, solve
x, y = symbols("x y")
f1=-1+3*x**2-2*y**2
f2=-8-2*x+x**2+2*y+y**2
M=[[(sp.diff(f1,x)),(sp.diff(f1,y))],[(sp.diff(f2,x)),(sp.diff(f2,y))]]
W=[[(2*sp.diff(f1,x)),(2*sp.diff(f2,x))],[(2*sp.diff(f1,y)),(2*sp.diff(f2,y))]]
print("")
print(M)
print("")
print(W)
print("")
```

```python
def eva(i,x,y):
  g1=-1+3*x**2-2*y**2
  g2=-8-2*x+x**2+2*y+y**2
  g3=g1**2+g2**2
  if i==1:
    return float(g1)
  if i==2:
    return float(g2)
  if i==3:
    return float(g3)
def eva2(M,i,j):
  v1=M[i][j]
  a=lambda x,y:v1
  v=a(x,y)
  return float(v)
```

```python
#2x2
import numpy as np
#Intervalo
x=1.5
y=1.5
#No olvidar Matriz
M=[[12*x, 4*x - 4], [-8*y, 4*y + 4]]
M3=np.array([eva(1,x,y),eva(2,x,y)])
M2=np.array([[eva2(M,0,0), eva2(M,0,1)], [eva2(M,1,0), eva2(M,1,1)]])
X=M2.dot(M3)
z0=(X[0]**2+X[1]**2)**0.5
z1=X[0]/z0
z2=X[1]/z0
aa=0
g1=eva(3,x,y)
a3=1
g3=eva(3,x-a3*z1,y-a3*z2)
while g3>g1:
  a3=a3/2
  g3=eva(3,x-a3*z1,y-a3*z2)
a2=a3/2
g2=eva(3,x-a2*z1,y-a2*z2)
h1=(g2-g1)/(a2-aa)
h2=(g3-g2)/(a3-a2)
h3=(h2-h1)/(a3-aa)
Pa=0.5*(a2-(h1/h3))
xf=x-Pa*z1
yf=y-Pa*z2
print("gradiente g(x0)",X)
print("Z0=",z0)
print("Z=",z1,z2)
print("a1=",aa,"     g1=",g1,"              h1=",h1)
print("a2=",a2,"  g2=",g2,"         h2=",h2)
print("a3=",a3,"   g3=",g3,"         h3=",h3)
print("P(a)=",Pa)
print("----------------------------")
print(xf)
print(yf)
```
