```python
import numpy as np

# Definición de las funciones
def f1(x, y):
    return x*2 + y*2 - 1

def f2(x, y):
    return x*2 - y*2 + 0.5

# Definición del Jacobiano
def jacobian(x, y):
    return np.array([[2*x, 2*y],
                     [2*x, -2*y]])

# Método de Newton utilizando matriz aumentada
def newton_method(x0, y0, tol=1e-10, max_iter=1000):
    xn, yn = x0, y0
    for i in range(max_iter):
        # Vector de funciones evaluado en (xn, yn)
        F = np.array([f1(xn, yn), f2(xn, yn)])

        # Jacobiano evaluado en (xn, yn)
        J = jacobian(xn, yn)

        # Matriz aumentada [J | -F]
        aug_matrix = np.hstack([J, -F.reshape(-1, 1)])

        # Resolución del sistema lineal usando eliminación de Gauss
        delta, residuals, rank, s = np.linalg.lstsq(J, -F, rcond=None)
        xn += delta[0]
        yn += delta[1]

        # Imprimir los valores actuales de x e y
        print(f"Iteración {i}: x = {xn}, y = {yn}")

        # Verificar convergencia
        if np.linalg.norm(delta) < tol:
            return xn, yn, i

    return xn, yn, i  # Devuelve la última iteración si no converge

# Punto inicial
x0, y0 = (1.510753, 1.808246)

# Llamada al método de Newton
x_sol, y_sol, iteraciones = newton_method(x0, y0)
print(f"Solución: x = {x_sol}, y = {y_sol}, en {iteraciones} iteraciones.")
```
