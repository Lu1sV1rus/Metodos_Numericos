```python
import math
#Funcion
#x(t)=-(g/2w^2)[((e^wt-e^-wt)/2))-senwt]
#Valores
t=1 #x=1.7, g=32.17
#Equivalencia
#1.7=-(32.17/2w^2)[sinh(w)-senw]
#(32.17/2w^2)[sinh(w)-senw]+1.7=0

def g(w):
  s=32.17/2
  s1=(w**2)
  s2=s/s1
  s3=math.sinh(w)-math.sin(w)
  s4=(s2*s3)+1.7
  return s4

#f=((32.17/2w^2)[sinh(w)-senw]+1.7)dw
#f'=(32.17/2)(1/w^4)((w^2)[cosh(w)-cosw]-[sinh(w)-senw](2w))
def g1(w):
  s=32.17/2
  s1=(w**2)
  a1=math.cosh(w)-math.cos(w)
  k1=s1*a1
  s2=2*w
  a2=math.sinh(w)-math.sin(w)
  k2=-s2*a2
  s4=k1+k2
  a3=(w**4)
  s5=s4/a3
  s6=(s*s5)
  return s6

ep=1e-5
p=12
n=0
print("N | W       | f(w)=     | f'(w)=   | (g(p)/g1(p))| p-(g(p)/g1(p))")
print("{} | {}      | {:.6f} | {:.4f} | {:.4f} | {:.4f}".format(n,p,g(p),g1(p),(g(p)/g1(p)),p-(g(p)/g1(p))))
#Metodo de Newton
while abs(g(p))>=ep:
  p=p-(g(p)/g1(p))
  n=n+1
  if n>=200:
    break
  print("{} | {:.4f} | {:.6f} | {:.4f} | {:.4f} | {:.4f}".format(n,p,g(p),g1(p),(g(p)/g1(p)),p-(g(p)/g1(p))))
print("Con {} iteraciones obtenemos que w={:.6f} que es raiz {}".format(n,p,g(p)))
print("Al evaluar w en la funcion",-(32.17/(2*(p**2)))*(math.sinh(p)-math.sin(p)))
#d(angulo)/dt=w<0
#angulo=wdt<0dt
#angulo=wt<0
print("El angulo {:.4f}º".format(math.degrees(p*t)))
```
