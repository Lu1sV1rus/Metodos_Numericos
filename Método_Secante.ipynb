```python
import math
def f(x):
    g=-(x**3)-math.cos(x)
    return g

p0=4
p1=-2
q0=f(p0)
q1=f(p1)
print("0 | {} | {} |".format(p0,f(p0)))
print("1 | {} | {}  |".format(p1,f(p1)))
ep=1e-15
for i in range(100):
    pf=p1-((q1*(p1-p0))/(q1-q0))
    p0=p1
    p1=pf
    q0=f(p0)
    q1=f(p1)
    if abs(f(pf))<ep:
       break
    print("{} | {:.15f} | {}  |".format(i+2,p1,f(p1)))
print("La raíz es {}".format(p1))
```
