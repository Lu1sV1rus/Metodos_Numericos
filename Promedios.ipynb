```python
import math
#Formula de 3pts
x=2.0
h=0.1
vi=0

def fun(x):
  fx=x*((math.e)**x)
  return fx

def delan(x,h):
  gx=(1/(2*h))*(-3*fun(x)+4*fun(x+h)-fun(x+(2*h)))
  return gx

def medio(x,h):
  gx=(1/(2*h))*(-fun(x-h)+fun(x+h))
  return gx

def atras(x,h):
  gx=(1/(2*h))*(3*fun(x)-4*fun(x-h)+fun(x-(2*h)))
  return gx

print("Formula hacia adelante =",delan(x,h))
print("error =",abs(vi-delan(x,h)))
print("------")
print("Formula medio =",medio(x,h))
print("error =",abs(vi-medio(x,h)))
print("------")
print("Formula hacia atras =",atras(x,h))
print("error =",abs(vi-atras(x,h)))
```

```python
import math
#Formulas de derivadas
x=1.3
h=0.1
vi=0

def fun(x):
  fx=3*x*((math.e)**x)-math.cos(x)
  return fx

def medio(x,h):
  gx=(1/(12*h))*(fun(x-(2*h))-8*fun(x-h)+8*fun(x+h)-fun(x+(2*h)))
  return gx

def doble(x,h):
  gx=(1/(h**2))*(fun(x+h)+fun(x-h)-2*fun(x))
  return gx

def triple(x,h):
  gx=(3/(h**3))*(fun(x+h)-fun(x-h)-(2*h*(medio(x,h))))
  return gx

print("Derivada =",medio(x,h))
print("error =",abs(vi-medio(x,h)))
print("------")
print("Doble Derivada =",doble(x,h))
print("error =",abs(vi-doble(x,h)))
print("------")
print("Tripe Derivada =",triple(x,h))
print("error =",abs(vi-triple(x,h)))
```

```python
#Richardson
import math
import numpy as np

x=1.05
h=0.4
n=4
vi=0

def fun(x):
  fx=(2**x)*math.sin(x)
  return fx

def medio(x,h):
  gx=(1/(2*h))*(-fun(x-h)+fun(x+h))
  return gx

i=1
R=[]
c2=n
while i<n+1:
  h2=h/(2**(i-1))
  i=i+1
  list.append(R,medio(x,h2))
  print(R[i-2])
for k in range(2,n+1):
  print("------")
  c2=c2-1
  for j in range(k,n+1):
    c1=len(R)
    list.append(R,(((4**(k-1))*R[c1-c2])-R[c1-c2-1])/((4**(k-1))-1))
    print(R[c1])

print("------")
print("f(x)=",vi)
print("------")
vf=(R[len(R)-1])
print("error",abs(vi-vf))
```

```python
#Reglas Compuestas (Integrales)
import math
import numpy as np

a=0
b=2*math.pi
n=4
vi=-2*math.pi
def fun(x):
  fx=x*math.sin(x)
  return fx

def trapeciocomp(a,b,n):
  i=1
  w=0
  h=(b-a)/n
  while i<n:
    w=w+fun(a+(i*h))
    i=i+1
  gx=(((1/2)*h)*(fun(a)+fun(b)+2*w))
  return gx

def simpsmcomp(a,b,n):
  i=1
  j=2
  w=0
  v=0
  h=(b-a)/(2*n)
  while i<(2*n):
    w=w+fun(a+(i*h))
    i=i+2
  while j<(2*n):
    v=v+fun(a+(j*h))
    j=j+2
  gx=(((1/3)*h)*(fun(a)+fun(b)+4*w+2*v))
  return gx

def mediocomp(a,b,n):
  j=1
  w=0
  h=(b-a)/(n)
  while j<=(n):
    w=w+fun((a+(j-(1/2))*h))
    j=j+1
  gx=(h*(w))
  return gx

print("f(x) =",vi)
print("------")
print("Trapecio Compuesto =",trapeciocomp(a,b,n))
print("error",abs((vi)-trapeciocomp(a,b,n)))
print("------")
print("Simpsm Compuesto =",simpsmcomp(a,b,n))
print("error",abs((vi)-simpsmcomp(a,b,n)))
print("------")
print("Punto Medio Compuesto =",mediocomp(a,b,n))
print("error",abs((vi)-mediocomp(a,b,n)))
```
