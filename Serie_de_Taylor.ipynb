```python
import math
import sympy as sp
# variables simbolicas
x , y ,z = sp.symbols('x y z')
e=sp.exp(-x)

# Definicion de una funcion simbolica
f = (sp.cos(x))*(e)
f1 = sp.diff(f,x)
f2 = sp.diff(f1,x)
f3 = sp.diff(f2,x)

fx  = sp.lambdify(x,f)
f1x = sp.lambdify(x,f1)
f2x = sp.lambdify(x,f2)
f3x = sp.lambdify(x,f3)
```

```python
#Con n=2 y x=0
x=0                                              # x valuada en 0
f=0.01
a=(math.exp(-x)*math.cos(x))
b=(-math.exp(-x)*(math.cos(x)+math.sin(x)))
c=(2*math.exp(-x)*(math.sin(x)))
d=(-2*(math.exp(-x)*(math.sin(x)-math.cos(x))))
df=(-2*(math.exp(-f)*(math.sin(f)-math.cos(f))))
print("Valores de las funciones valuadas en ",x)
print(a)
print(b)
print(c)
print(d)

#   1         2         3               Mi serie de Taylor con valores en x=0
P2=(a*(1))+(b*(1)*f)+(c*(1/2)*(f**2))
#   4                                   Mi ultimo valor dependiente de f=0.01
P24=(df*(1/6)*f**3)                     # Mi P2 en la posicion 4, P24
print("Mi residou de orden 2 es ",P24)
print("El error de truncamiento es")
print(P2," + ",P24)
P2total=P2+P24
print("El valor abs del error es")      #Valor abs del error
error=abs(math.cos(f)-P2)
print(error)

if P2total>=math.cos(f):                # Mi limite de error
  print("El error real esta por encima de mi limite")
else:
  print("El error real esta por debajo de mi limite")
```
