# Teorema de Taylor

Sea $f : [a,b] \rightarrow B \subset \mathbb{R}$ una función $n+1$ veces deribable en $(a,b)$ y $x_{0} \in [a,b]$, entonces para cada $x \in [a,b]$ existe $\xi(x)$ entre $x_{0}$ y $x$ tal que
\begin{equation}
f(x) = P_{n}(x) + R_{n}(x)
\end{equation}
donde
\begin{equation}
P_{n}(x) = \sum_{k=0}^{n}\frac{f^{(k)}(x_{0})}{k!}(x-x_{0})^{k}
\end{equation}
y
\begin{equation}
R_{n}(x) = \frac{f^{n+1}(\xi(x))}{(n+1)!}(x-x_{0})^{n+1}
\end{equation}
