# Repaso de calculo: Sucesiones

Definición: Límite de una sucesión.

Sea $\{x_{n}\}^{+\infty}_{n=1}$ una sucesion de números reales. Essta sucesión converge a $l$  si y solo si.
para todo $\varepsilon > 0$ existe un $n_{0} \in \mathbb{N}$ tal que $\left|x_{n} - l\right|<\varepsilon$ para todo $n \geq n_{0} $

# Ejemplo
considere la sucesion $a_{1} = \sqrt{2}$ , $a_{n+1} = \sqrt{2 a_{n}}$, para $n = 1,2,3,\dots$

```python
# Ejemplo
# Este es un comentario en python
# Importamos la libreria math
# para hacer uso de las liberias matematicas
import math

# Definiremos la sucesion en virtud a su recurrencia
def an(n):
    # condicional para el valor inicial
    if n == 1:
        a = math.sqrt(2.0)
    else :
        a = math.sqrt(2.0*an(n-1))
    return a

# el limite de la sucesion es
l = 2.0

# El valor de epsilon (exactitud, accurancy)
epsilon = 1.0e-3

n = 1
a = an(n)
while abs(a - l) >= epsilon :
    print('n = {}, an = {:.4f}'.format(n,a))
    a = an(n)
    n += 1
print('n = {}, an = {:.4f}'.format(n,a))
print('--------------------------------')
print('El valor de n0 es {}'.format(n))
```
